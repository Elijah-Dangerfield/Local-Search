```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

```python
datax = np.loadtxt('data-2.in')
datax
```

```python
datay = np.loadtxt('data-2.out')
```

```python
greedy = np.loadtxt('search.out')
```

```python
# 3D Plot - All scatter
fig = plt.figure()
ax = fig.gca(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
ax.scatter(datax[:,0],datax[:,1],datay[:,0],s=2)
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0.0, vmax=1.0)
ax.scatter(greedy[:,0],greedy[:,1],greedy[:,2],color=cmap(norm(greedy[:,2])))
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("G(X)")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```
