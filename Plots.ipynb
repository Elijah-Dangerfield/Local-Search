```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline
```

# 1D Example (second dimension is G(X))

```python
greedy = np.loadtxt("greedy-0-1-1.txt")
greedy
```

```python
datax = np.loadtxt("data-1.in")
datax
```

```python
datay = np.loadtxt("data-0-1-1.txt")
datay
```

```python
plt.plot(datax,datay[:,0])
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0.0, vmax=1.0)
plt.scatter(greedy[:,0],greedy[:,1],color=cmap(norm(greedy[:,1])))
plt.ylabel('G(X)')
plt.xlabel('X')
plt.show()
```

# 2D Example (third dimension is G(X))

```python
greedy = np.loadtxt("greedy-0-2-1.txt")
greedy
```

```python
datax = np.loadtxt("data-2.in")
datax
```

```python
datay = np.loadtxt("data-0-2-1.txt")
datay
```

```python
# 3D Plot - All scatter
fig = plt.figure()
ax = fig.gca(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
ax.scatter(datax[:,0],datax[:,1],datay[:,0],s=2)
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0.0, vmax=1.0)
ax.scatter(greedy[:,0],greedy[:,1],greedy[:,2],color=cmap(norm(greedy[:,2])))
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("G(X)")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```

```python
# 3D Plot - surface + scatter
fig = plt.figure()
ax = fig.gca(projection='3d')
fig.set_figwidth(10)
fig.set_figheight(8)
X = datax[:,0].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Y = datax[:,1].reshape((int(np.sqrt(datax.shape[0])),int(np.sqrt(datax.shape[0]))))
Z = datay[:,0].reshape(X.shape)
ax.plot_surface(X, Y, Z)
cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0.0, vmax=1.0)
ax.scatter(greedy[:,0],greedy[:,1],greedy[:,2],color=cmap(norm(greedy[:,2])),s=10)
ax.set_xlabel("x0")
ax.set_ylabel("x1")
ax.set_zlabel("G(X)")
ax.view_init(45,-45) # Note that angles may need to be changed to visualize properly...
```
